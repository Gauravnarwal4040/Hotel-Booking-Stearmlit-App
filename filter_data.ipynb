```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data\hotel_booking.csv")
```

```python
df.info()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df['children'].fillna(df['children'].mode()[0],inplace=True)
df['country'].fillna(df['country'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(columns=['name','email','phone-number','credit_card','agent','company'],inplace=True)
```

```python
df['arrival_month_num'] = pd.to_datetime(df['arrival_date_month'],format = "%B").dt.month
```

```python
df['assigned_room_status'] = df['assigned_room_type'] == df['reserved_room_type']
```

```python
with open("data/clean_hotel_booking.csv","w") as file:
    df.to_csv(file,index = False)
```
